```python
# confusion matrix

def plot_confusion_matrix(cm, classes = None, title='Confusion matrix'):
    """Plots a confusion matrix."""
    if classes is not None:
        sns.heatmap(cm, xticklabels=classes, yticklabels=classes, vmin=0., vmax=1., annot=True, annot_kws={'size':50})
    else:
        sns.heatmap(cm, vmin=0., vmax=1.)
    plt.title(title)
    plt.ylabel('True Label')
    plt.xlabel('Predicted Label')
```

```python
# Visualization CM
from sklearn.metrics import confusion_matrix


cm = confusion_matrix(y_train, y_pred)
cm_perc = cm / cm.sum(axis=1).reshape(-1,1)

plot_confusion_matrix(cm_perc, classes = log_reg.classes_, title='Confusion matrix')
```

```python
# calc FP, FN, TP, TN
# ini array dari CM aja

fp = cm.sum(axis=0) - np.diag(cm)
fn = cm.sum(axis=1) - np.diag(cm)
tp = np.diag(cm)
tn = cm.sum() - (fp+fn+tp)

# TP rate
tpr = tp / (tp+fn)
print('TP Rate = ', tpr)

# Positive predictive value
ppv = tp / (tp+fp)
print('Precision = ', ppv)

# False alarm rate
fpr = fp / (fp+tn)
print('FP Rate = ', fpr)

# False negative
fnr = fn / (fn+tp)
print('FN Rate = ', fnr)

# Average
print('')
print('Average TPR = ', tpr.sum()/2)
print('Average Precision = ', ppv.sum()/2)
print('Average FPR = ', fpr.sum()/2)
print('Average FNR = ', fnr.sum()/2)
```

```python
import keras_tuner as kt
import tensorflow as tf
from tensorflow import keras
```

```python
def build_model(hp):
    model = keras.Sequential()
    model.add(keras.layers.Dense(units = 16, activation = 'relu', input_dim =111))
    # model.add(keras.layers.Dropout(0.2))
    # model.add(keras.layers.Dense(units = 16, activation = 'relu'))
    for i in range(hp.Int("num_layers", 1, 2)):
        model.add(keras.layers.Dense(hp.Choice('units', [16, 32]),activation='relu'))
    if hp.Boolean("dropout"):
        model.add(keras.layers.Dropout(rate=0.25))
    model.add(keras.layers.Dense(units = 1, activation = 'sigmoid'))
    hp_learning_rate = hp.Choice('learning_rate', values=[1e-2, 1e-3, 1e-4])
    # hp_learning_rate = 1e-2
    model.compile(optimizer=keras.optimizers.Adam(learning_rate=hp_learning_rate),
                  loss=keras.losses.BinaryCrossentropy(from_logits=True),
                  metrics=['accuracy'])
    return model
```

```python
tuner = kt.Hyperband(
    build_model,
    objective='val_accuracy',
    max_epochs=10,
    factor=3,
    directory='check',
    overwrite=True)
```

```python
stop_early = tf.keras.callbacks.EarlyStopping(monitor='val_loss')
```

```python
tuner.search(X_train, y_train, epochs=50, validation_data=(X_test, y_test), callbacks=[stop_early])
```

```python
best_hps=tuner.get_best_hyperparameters(num_trials=1)[0]
print(f"""
The hyperparameter search is complete. 
The optimal number of units in the first densely-connected layer is,
{best_hps.get('units')}
the optimal learning rate for the optimizer is,
{best_hps.get('learning_rate')}
the optimal number of layers for the optimizer is
{best_hps.get('num_layers')}
the optimal dropouts for the optimizer is,
{best_hps.get('dropout')}
""")
```
