```python
# confusion matrix

def plot_confusion_matrix(cm, classes = None, title='Confusion matrix'):
    """Plots a confusion matrix."""
    if classes is not None:
        sns.heatmap(cm, xticklabels=classes, yticklabels=classes, vmin=0., vmax=1., annot=True, annot_kws={'size':50})
    else:
        sns.heatmap(cm, vmin=0., vmax=1.)
    plt.title(title)
    plt.ylabel('True Label')
    plt.xlabel('Predicted Label')
```

```python
# Visualization CM
from sklearn.metrics import confusion_matrix


cm = confusion_matrix(y_train, y_pred)
cm_perc = cm / cm.sum(axis=1).reshape(-1,1)

plot_confusion_matrix(cm_perc, classes = log_reg.classes_, title='Confusion matrix')
```

```python
# calc FP, FN, TP, TN
# ini array dari CM aja

fp = cm.sum(axis=0) - np.diag(cm)
fn = cm.sum(axis=1) - np.diag(cm)
tp = np.diag(cm)
tn = cm.sum() - (fp+fn+tp)

# TP rate
tpr = tp / (tp+fn)
print('TP Rate = ', tpr)

# Positive predictive value
ppv = tp / (tp+fp)
print('Precision = ', ppv)

# False alarm rate
fpr = fp / (fp+tn)
print('FP Rate = ', fpr)

# False negative
fnr = fn / (fn+tp)
print('FN Rate = ', fnr)

# Average
print('')
print('Average TPR = ', tpr.sum()/2)
print('Average Precision = ', ppv.sum()/2)
print('Average FPR = ', fpr.sum()/2)
print('Average FNR = ', fnr.sum()/2)
```

```python
# Looping over the parameter

C_List = np.geomspace(1e-5, 1e5, num=30)
CA = []
Logarithmic_Loss = []

for c in C_List:
    log_reg2 = LogisticRegression(random_state=23, solver = 'newton-cg', C = c, max_iter=5000)
    log_reg2.fit(X_train,y_train)
    score = log_reg2.score(X_test, y_test)
    CA.append(score)
    print(f"the CA of C parameter {c} is {score}")
    pred_proba_t = log_reg2.predict_proba(X_test)
    log_loss2 = log_loss(y_test, pred_proba_t)
    Logarithmic_Loss.append(log_loss2)
    print(f"Log loss of C parameter {c} is {log_loss2}")
    print("")
```

```python
CA2 = np.array(CA).reshape(30)
Logarithmic_Loss2 = np.array(Logarithmic_Loss).reshape(30,)

outcomes = zip(C_List, CA2, Logarithmic_Loss2)

df_outcomes = pd.DataFrame(outcomes, columns=["C_List", "CA2", "Logarithmic_Loss2"])
df_outcomes
df_outcomes.sort_values("Logarithmic_Loss2", ascending=True).reset_index()
```
